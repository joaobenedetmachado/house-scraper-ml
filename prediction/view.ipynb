```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/data.csv")

df.head()
```

```python
print("media dos valores")

print("media precos", df['preco'].mean())

print("devio padrão: ", df['preco'].std()) # desvio padrao
```

```python
sns.boxplot(x='nquartos', y='preco', data=df)
plt.title("Preço por Número de Quartos")
plt.xlabel("Número de Quartos")
plt.ylabel("Preço")
plt.show()
```

```python
sns.scatterplot(x='nquartos', y='nbanheiros', size='preco', hue='preco', data=df, palette='viridis', alpha=0.7)
plt.title("Relação entre Quartos e Banheiros (tamanho pelo preço)")
plt.xlabel("Quartos")
plt.ylabel("Banheiros")
plt.show()
```

```python
corr = df[['preco','nquartos','nbanheiros','ngaragem']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlação entre Variáveis")
plt.show()
```
