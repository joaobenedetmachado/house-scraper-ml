```python
import pandas as pd

df = pd.read_csv('../data/raw/raw.csv')

df.info()
```

```python
df.head()
```

```python
df['preco'] = df['preco'].str.replace(r'[^\d]', '', regex=True)

df.head()
```

remover os cifroes dos precos

```python
df.dropna()
df.replace(r'^\s*$', pd.NA, regex=True, inplace=True)
df.dropna(inplace=True)
```

remover os metros quadrados

```python
df['area'] = df['area'].str.replace(r'[^\d,.]', '', regex=True)

df.head()
```

retirar os +

```python
df['nquartos'] = df['nquartos'].astype(str).str.replace('+', '', regex=False)
df['nbanheiros'] = df['nbanheiros'].astype(str).str.replace('+', '', regex=False)
df['ngaragem'] = df['ngaragem'].astype(str).str.replace('+', '', regex=False)


df.head()
```

salvar

```python
df.to_csv('../data/processed/data.csv', index=False)
```
